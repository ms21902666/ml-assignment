```python
X_ind = avocado['4770']
y_dep = avocado['Total Volume']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_ind, y_dep, train_size = 0.7,
test_size = 0.3, random_state = 100)
```

```python
import statsmodels.api as stats_mod
```

```python
X_train_sm = stats_mod.add_constant(X_train)
```

```python
lin_reg = stats_mod.OLS(y_train, X_train_sm).fit()
lin_reg.params
```

```python
lin_reg.summary()
```

```python
plott.scatter(X_train, y_train)
plott.plot(X_train, 2.043e+05 + 27.9131*X_train, 'r')
plott.show()
```

```python
y_train_pred = lin_reg.predict(X_train_sm)
res = (y_train - y_train_pred)
res
```

```python
fig = plott.figure()
sns.distplot(res, bins = 15)
plott.title('Error Terms', fontsize = 15)
plott.xlabel('y_train - y_train_pred', fontsize = 15)
plott.show()
```

```python
plott.scatter(X_train,res)
plott.show()
```

```python
X_test_sm = stats_mod.add_constant(X_test)
y_test_pred = lin_reg.predict(X_test_sm)
y_test_pred
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
plott.scatter(X_test, y_test)
plott.plot(X_test, y_test_pred, 'r')
plott.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = np.arange(10).reshape((5, 2)), range(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.7,
test_size = 0.3, random_state = 100)
X_train, X_test, y_train, y_test
```

```python
X_train.shape
X_train = X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print("Intercept :",lm.intercept_)
print('Slope :',lm.coef_)
```

```python
y_train_pred = lm.predict(X_train)
y_test_pred = lm.predict(X_test)
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```
